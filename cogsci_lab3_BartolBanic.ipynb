2. a) Find the mean, median and average scores in mathematics, reading and writing. <br>
   b) Find the mean, median and average scores for males and females.<br>
   c) What is the coefficient of variation of the scores? <br>

```python
means = student_data[["MathScore", "ReadingScore", "WritingScore"]].mean()
medians = student_data[["MathScore", "ReadingScore", "WritingScore"]].median()
modes = student_data[["MathScore", "ReadingScore", "WritingScore"]].mode()

means_m = student_data.loc[student_data["Gender"]== "male", ["MathScore", "ReadingScore", "WritingScore"]].mean()
medians_m = student_data.loc[student_data["Gender"]== "male", ["MathScore", "ReadingScore", "WritingScore"]].median()
modes_m = student_data.loc[student_data["Gender"]== "male", ["MathScore", "ReadingScore", "WritingScore"]].mode()

means_f = student_data.loc[student_data["Gender"]== "female", ["MathScore", "ReadingScore", "WritingScore"]].mean()
medians_f = student_data.loc[student_data["Gender"]== "female", ["MathScore", "ReadingScore", "WritingScore"]].median()
modes_f = student_data.loc[student_data["Gender"]== "female", ["MathScore", "ReadingScore", "WritingScore"]].mode()

coef_var_Math = student_data["MathScore"].std()/student_data["MathScore"].mean() * 100
coef_var_Read = student_data["ReadingScore"].std()/student_data["ReadingScore"].mean() * 100
coef_var_Write = student_data["WritingScore"].std()/student_data["WritingScore"].mean() * 100
coef_var_df = pd.DataFrame({"Coefficient of Variation": [coef_var_Math, coef_var_Read, coef_var_Write]}, index=["MathScore", "ReadingScore", "WritingScore"])

describ_df = pd.DataFrame({"Mean": means, "Median": medians, "Mode": modes.iloc[0].values, "Mean (Male)": means_m, "Median (Male)": medians_m, "Mode (Male)": modes_m.iloc[0].values, "Mean (Female)": means_f, "Median (Female)": medians_f,  "Mode (Female)": modes_f.iloc[0].values,})
descriptives = pd.concat([describ_df, coef_var_df], axis=1)
descriptives
```

3. a) Use a statistical test to test the hypothesis:<br>
H0: the mean mathematical scores of females and males are equal.



```python
from scipy import stats
from scipy.stats import shapiro
```

```python
stats.levene(student_data.loc[(student_data["Gender"]=="female"),"MathScore"], student_data.loc[(student_data["Gender"]=="male"),"MathScore"])
```

```python
#Since the Levene test shows that there is no significant evidence to suggest that the variances are different
#across the groups, a standard independent 2 sample t-test is performed.
stat_ms, p_ms = stats.ttest_ind(student_data.loc[(student_data["Gender"]=="female"),"MathScore"], student_data.loc[(student_data["Gender"]=="male"),"MathScore"])
print('stat=%.3f, p=%.3f' % (stat_ms, p_ms))
```

4. a) Use a statistical test to test the hypothesis:<br>
   H0 Gender did not affect the writing scores.<br>
   H1 Gender affected the writing score.<br>

```python
stats.levene(student_data.loc[(student_data["Gender"]=="female"),"WritingScore"], student_data.loc[(student_data["Gender"]=="male"),"WritingScore"])
```

```python
#Applying the same logic as above, a standard independent 2 sample t-test is performed.
stat_ws, p_ws = stats.ttest_ind(student_data.loc[(student_data["Gender"]=="female"),"WritingScore"], student_data.loc[(student_data["Gender"]=="male"),"WritingScore"])
print('stat=%.3f, p=%.3f' % (stat_ws, p_ws))
```

5. a) Statisticaly test the distribuitions of female and male scores across the three scores columns.<br>
b) Plot the distributions for each of the separate scores in both genders. <br>

```python
Scores = ["MathScore", "WritingScore", "ReadingScore"]
for score in Scores:
    stat1, p1 = shapiro(student_data.loc[student_data["Gender"]=="male", score])
    print(f'Male - {score}: stat={stat1:.3f}, p={p1:.3f}')
    if p1 > 0.05:
        print('Probably Gaussian')
    else:
        print('Probably not Gaussian')
    stat2, p2 = shapiro(student_data.loc[student_data["Gender"]=="female", score])
    print(f'Female - {score}: stat={stat2:.3f}, p={p2:.3f}')
    if p2 > 0.05:
        print('Probably Gaussian')
    else:
        print('Probably not Gaussian')
```

```python
fig, ax = plt.subplots(3,2, figsize=(30,30))
mask_m = student_data["Gender"]=="male"
Scores = ["MathScore", "WritingScore", "ReadingScore"]
for i, score in zip(range(3), Scores):
    ax[i,0].hist(student_data.loc[mask_m, score], bins=30, range=(0,100))
    ax[i,0].set_xlabel(score)
    ax[i,0].set_ylabel("Score frequency")
    ax[i,0].set_title(f"Male - {score}")
    ax[i,1].hist(student_data.loc[~mask_m, score], bins=30, range=(0,100))
    ax[i,1].set_xlabel(score)
    ax[i,1].set_ylabel("Score frequency")
    ax[i,1].set_title(f"Female - {score}")
plt.show()
```
